```python
# Splitting data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,label,test_size = 0.2, random_state=128)
```

```python
# Sacaling data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python

#Variabes to calculate sigmoid function
y_pred = []
len_x = len(X_train[0])
w = []
b = 0.2
print(len_x)
```

```python
def sigmoid(z):
 return 1/(1 + np.exp(-z))
```

```python
def log_loss(predicted,actual):
   epsilon = 1e-15
   predicted = [max(epsilon,i) for i in predicted] 
   predicted = [min(i,1-epsilon) for i in predicted]
   predicted_new = np.array(predicted)
   x = -np.mean(actual * np.log(predicted_new) + (1-actual) * np.log(1-predicted_new))
   return x
```

```python
def gradients(X, y, y_hat):
    m = X.shape[0]
    dw = (1/m)*np.dot(X.T, (y_hat - y))
    db = (1/m)*np.sum((y_hat - y)) 
    return dw, db
```

```python
def normalize(X):
  m, n = X.shape
  for i in range(n):
    X = (X - X.mean(axis = 0)) / X.std(axis = 0)
  return X
```

```python
def train(X, y, bs, epochs, lr):
    m, n = X.shape
    w = np.zeros((n,1))
    b = 0

    y = y.values.reshape(m,1)
    x = normalize(X)

    losses = []

    for epoch in range(epochs):
        for i in range((m-1)//bs + 1):
            start_i = i*bs
            end_i = start_i + bs
            xb = X[start_i:end_i]
            yb = y[start_i:end_i]
            y_hat = sigmoid(np.dot(xb, w) + b)
            dw, db = gradients(xb, yb, y_hat)
            w -= lr*dw
            b -= lr*db
        l = log_loss(y, sigmoid(np.dot(X, w) + b))
        losses.append(l)
    return w, b, losses
```

```python
def prediction_function(inputs):
  x = normalize(inputs)
  preds = sigmoid(np.dot(inputs, w) + b)
  pred_class = []    
  pred_class = [1 if i > 0.5 else 0 for i in preds]  
  return np.array(pred_class)
```

```python
w, b, l = train(X_train, y_train, bs=100, epochs=1000, lr=0.01)
```

```python
from sklearn.metrics import accuracy_score

y_predict = prediction_function(X_test)
print(accuracy_score(y_test, y_predict))
```
